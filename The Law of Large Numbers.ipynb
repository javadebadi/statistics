# The Law of Large Numbers

Let $X_i$ for $i=1,...,N$ be iid random variables with poplulation mean $\mu$. The sample mean 
$$\bar{X}_N = \frac{\sum_{i=1}^N X_i}{N}$$
is itself a random variable and thus has its own distirbution.

The law of large numbers (LLN) states that the $\bar{X}_N$ converges in probability to $\mu$.
In addition we know that the sample mean is an unbiased estimator of the population mean. Thus, we say that sample variance is an __unbiased__ and __consistent__ estimator of the population mean. 

In this lecture, we want to test the validity of LLN.

## Example : Die roll


```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def distributions_example_sample_mean_and_sample_variance_generated_die_data(size=1):
    """
    input: size=sampel size
    output: 100000 data which average of size times rolling a die
    a function that generate 100000 data for die roll with sample size = size
    """

    die_min = 1    # minimum value a die can return
    die_max = 6    # maximum value a die can return
    n_sim = 100000   # number of times we run the simulation (rolling a die)
    data=[]
    for item in xrange(n_sim):
        temp = np.random.randint(low=die_min, high=die_max+1, size=size).sum()*(1.0/size)
        data.append(temp)
    return data
        
def distributions_example_sample_mean_and_sample_variance_make_plot_title(size):
    """
    set titles for the plot for die problem
    
    """
    plt.title("N (sample size) = "+str(size))
    plt.xlabel("die value")
    plt.ylabel("frequency (normalized)")
    plt.xlim([0,7])
    return 
        
def distributions_example_sample_mean_and_sample_variance_plot_for_die_samples(sizes=[1,2,6,30], n_bins=30):

    
    sub_plots_list=[221,222,223,224]
    fig = plt.figure(figsize=(10,10))
    counter = 0
    for size in sizes:
        data = distributions_example_sample_mean_and_sample_variance_generated_die_data(size=size)       
        plt.subplot(sub_plots_list[counter])
        #ax = fig.add_subplot(sub_plots_list[counter])
        distributions_example_sample_mean_and_sample_variance_make_plot_title(size)
        counter += 1
        # plot the histogram
        sns.distplot(data,bins=n_bins,kde=False)
    plt.tight_layout(h_pad=2.0,w_pad=2.0)
```

```python
distributions_example_sample_mean_and_sample_variance_plot_for_die_samples(n_bins=30)
```
